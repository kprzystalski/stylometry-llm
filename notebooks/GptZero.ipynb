## Wikipedia-like prompt

```python
import pandas as pd
import random

llama2_wiki = pd.read_csv('stylo_wiki_llama2.csv')
llama3_wiki = pd.read_csv('stylo_wiki_llama3.csv')
falcon_wiki = pd.read_csv('stylo_wiki_falcon.csv')
gpt3_wiki = pd.read_csv('stylo_wiki_gpt3.csv')
gpt4_wiki = pd.read_csv('stylo_wiki_gpt4.csv')
orca_wiki = pd.read_csv('stylo_wiki_orca.csv')
human = pd.read_csv('stylo_wiki_text.csv')

human_ids = random.sample(range(0, human.shape[0]), 100)
human_df = human.iloc[human_ids]['text']

llama2_wiki_ids = random.sample(range(0, llama2_wiki.shape[0]), 100)
llama2_wiki_df = llama2_wiki.iloc[llama2_wiki_ids]['text']

llama3_wiki_ids = random.sample(range(0, llama3_wiki.shape[0]), 100)
llama3_wiki_df = llama3_wiki.iloc[llama3_wiki_ids]['text']

falcon_wiki_ids = random.sample(range(0, falcon_wiki.shape[0]), 100)
falcon_wiki_df = falcon_wiki.iloc[falcon_wiki_ids]['text']

gpt3_wiki_ids = random.sample(range(0, gpt3_wiki.shape[0]), 100)
gpt3_wiki_df = gpt3_wiki.iloc[gpt3_wiki_ids]['text']

gpt4_wiki_ids = random.sample(range(0, gpt4_wiki.shape[0]), 100)
gpt4_wiki_df = gpt4_wiki.iloc[gpt4_wiki_ids]['text']

orca_wiki_ids = random.sample(range(0, orca_wiki.shape[0]), 100)
orca_wiki_df = orca_wiki.iloc[orca_wiki_ids]['text']
```

```python
import requests

url = "https://api.gptzero.me/v2/predict/text"


def checkAI(text):
    payload = { 
        "document": text,
    }
    headers = {
        "accept": "application/json",
        "x-api-key": "",
        "content-type": "application/json"
    }

    response = requests.post(url, json=payload, headers=headers)
    return response
```

```python
sum_data = pd.DataFrame()

print("HUMAN")

for i in human_wiki_df.index:
    doc = checkAI(human_wiki_df.loc[i])
    print(str(i))
    print(doc.json()['documents'][0]['predicted_class'])
    print(doc.json()['documents'][0]['confidence_score'])
    
    wikipedia_row = {'base_text': human_wiki_df.loc[i], 'model': 'human', 'prediction':  doc.json()['documents'][0]['predicted_class'],
                         'confidence': doc.json()['documents'][0]['confidence_score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('gptzero_wiki_human.csv')


sum_data = pd.DataFrame()

print("ORCA")

for i in orca_wiki_df.index:    
    doc = checkAI(orca_wiki_df.loc[i])
    print(str(i))
    print(doc.json()['documents'][0]['predicted_class'])
    print(doc.json()['documents'][0]['confidence_score'])
    
    wikipedia_row = {'base_text': orca_wiki_df.loc[i], 'model': 'orca', 'prediction':  doc.json()['documents'][0]['predicted_class'],
                         'confidence': doc.json()['documents'][0]['confidence_score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('gptzero_wiki_orca.csv')


sum_data = pd.DataFrame()

print("FALCON")

for i in falcon_wiki_df.index:
    doc = checkAI(falcon_wiki_df.loc[i])
    print(str(i))
    print(doc.json()['documents'][0]['predicted_class'])
    print(doc.json()['documents'][0]['confidence_score'])
    
    wikipedia_row = {'base_text': falcon_wiki_df.loc[i], 'model': 'falcon', 'prediction':  doc.json()['documents'][0]['predicted_class'],
                         'confidence': doc.json()['documents'][0]['confidence_score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('gptzero_wiki_falcon.csv')
```
