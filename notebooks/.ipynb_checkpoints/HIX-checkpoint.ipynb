```python
import pandas as pd
import random

llama2_wiki = pd.read_csv('stylo_wiki_llama2.csv')
llama3_wiki = pd.read_csv('stylo_wiki_llama3.csv')
falcon_wiki = pd.read_csv('stylo_wiki_falcon.csv')
gpt3_wiki = pd.read_csv('stylo_wiki_gpt3.csv')
gpt4_wiki = pd.read_csv('stylo_wiki_gpt4.csv')
orca_wiki = pd.read_csv('stylo_wiki_orca.csv')
human = pd.read_csv('stylo_wiki_text.csv')

human_ids = random.sample(range(0, human.shape[0]), 100)
human_df = human.iloc[human_ids]['text']

llama2_wiki_ids = random.sample(range(0, llama2_wiki.shape[0]), 100)
llama2_wiki_df = llama2_wiki.iloc[llama2_wiki_ids]['text']

llama3_wiki_ids = random.sample(range(0, llama3_wiki.shape[0]), 100)
llama3_wiki_df = llama3_wiki.iloc[llama3_wiki_ids]['text']

falcon_wiki_ids = random.sample(range(0, falcon_wiki.shape[0]), 100)
falcon_wiki_df = falcon_wiki.iloc[falcon_wiki_ids]['text']

gpt3_wiki_ids = random.sample(range(0, gpt3_wiki.shape[0]), 100)
gpt3_wiki_df = gpt3_wiki.iloc[gpt3_wiki_ids]['text']

gpt4_wiki_ids = random.sample(range(0, gpt4_wiki.shape[0]), 100)
gpt4_wiki_df = gpt4_wiki.iloc[gpt4_wiki_ids]['text']

orca_wiki_ids = random.sample(range(0, orca_wiki.shape[0]), 100)
orca_wiki_df = orca_wiki.iloc[orca_wiki_ids]['text']
```

```python
import requests
import time



url_submit = "https://bypass.hix.ai/api/hixbypass/v1/submit"
url_obtain = "https://bypass.hix.ai/api/hixbypass/v1/obtain"


def checkAI(text):
    payload_task = { 
        "input": text,
        "mode" : "Fast",
    }
    headers = {
        "Content-Type" : "application/json",
        "api-key": "",
    }
    
    task_id = requests.post(url_submit, json=payload_task, headers=headers).json()['data']['task_id']

    headers = {
        "Content-Type" : "application/json",
        "api-key": "",
    }
    res_pos = True
    while res_pos:
        response = requests.get(url_obtain+"?task_id="+task_id, headers=headers)
    
        if response.json()['err_msg'] == "Success":
            print(response.json()['err_msg'])
            if  response.json()['data']['subtask_status'] == "failed":
                res_pos = False
                break
            print(response.json()['data']['subtask_status'])
            if response.json()['data']['subtask_status'] == "completed":
                res_pos = False
                break
            else:
                time.sleep(2)
        else:
            print(response.json()['err_msg'])
            time.sleep(2)
    return {'prediction': response.json()['data']['detection_result'], 
            'score': response.json()['data']['detection_score']}
```

```python
text = "The list of Swedish royal consorts refers to the wives and husbands of the monarchs who have ruled Sweden throughout its history. The earliest recorded consort was Queen Enheduanna, the wife of King Erik VII, in the 12th century. She is known to have played a significant role in state affairs during her husband's reign. In the Middle Ages, consorts were often queens dowager or princesses who married into the Swedish royal family. The first queen consort to hold significant influence was Queen Margaret Valdemarsdottir, wife of King Magnus IV, who served as regent for her son-in-law during his minority. The 17th century saw a rise in consorts' influence, with notable examples including Queen Hedvig Eleonora and Queen Ulrika Eleonora. In the modern era, consorts have continued to play important roles in Swedish royal life, with notable figures including King Gustaf V's wife, Queen Victoria, and King Carl XVI Gustaf's wife, Queen Silvia. The current king of Sweden, Carl XVI Gustaf, is married to Queen Silvia, who has been a prominent supporter of her husband during his reign. Throughout its history, the list of Swedish royal consorts reflects the evolving role and influence of women in Swedish monarchy."
response = checkAI(text)
print(response)
```

```python
sum_data = pd.DataFrame()

print("HUMAN")

for i in human_df.index:
    doc = checkAI(human_df.loc[i])
    print(str(i))    
    print(str(doc['prediction']))
    print(str(doc['score']))
    wikipedia_row = {'base_text': human_df.loc[i], 'model': 'human', 'prediction':  doc['prediction'],
                         'confidence': doc['score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('hix_wiki_human.csv')
```

```python



sum_data = pd.DataFrame()

print("ORCA")

for i in orca_wiki_df.index:
    print(orca_wiki_df.loc[i])
    try:
        doc = checkAI(orca_wiki_df.loc[i])
    except:
        break
    print(str(i))    
    print(str(doc['prediction']))
    print(str(doc['score']))
    
    wikipedia_row = {'base_text': orca_wiki_df.loc[i], 'model': 'orca', 'prediction':  doc['prediction'],
                         'confidence': doc['score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('hix_wiki_orca.csv')


sum_data = pd.DataFrame()

print("FALCON")

for i in falcon_wiki_df.index:
    print(falcon_wiki_df.loc[i])
    try:
        doc = checkAI(falcon_wiki_df.loc[i])
    except:
        break
    print(str(i))    
    print(str(doc['prediction']))
    print(str(doc['score']))
    
    wikipedia_row = {'base_text': falcon_wiki_df.loc[i], 'model': 'falcon', 'prediction':  doc['prediction'],
                         'confidence': doc['score']}
        
    sum_data = pd.concat([sum_data, pd.DataFrame([wikipedia_row])], ignore_index=True)    
sum_data.to_csv('hix_wiki_falcon.csv')
```
