```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

# LLMs
gpt3_stylos = pd.read_csv('stylo_gpt3.csv')
gpt4_stylos = pd.read_csv('stylo_gpt4.csv')
llama2_stylos = pd.read_csv('stylo_llama2.csv')
llama3_stylos = pd.read_csv('stylo_llama3.csv')
falcon_stylos = pd.read_csv('stylo_falcon.csv')
orca_stylos = pd.read_csv('stylo_orca.csv')

# summarizers
wiki_sum_stylo = pd.read_csv('stylo_wiki_text.csv')
sumy_sum_stylo = pd.read_csv('stylo_sumy.csv')
ts_sum_stylo = pd.read_csv('stylo_ts.csv')
gensim_sum_stylo = pd.read_csv('stylo_gensim.csv')
bart_sum_stylo = pd.read_csv('stylo_bart.csv')

# wiki LLMs
wiki_gpt3_stylos = pd.read_csv('stylo_wiki_gpt3.csv')
wiki_gpt4_stylos = pd.read_csv('stylo_wiki_gpt4.csv')
wiki_llama2_stylos = pd.read_csv('stylo_wiki_llama2.csv')
wiki_llama3_stylos = pd.read_csv('stylo_wiki_llama3.csv')
wiki_falcon_stylos = pd.read_csv('stylo_wiki_falcon.csv')
wiki_orca_stylos = pd.read_csv('stylo_wiki_orca.csv')

# wiki
wiki_stylos = pd.read_csv('stylo_wiki.csv')

wiki_text_stylos = pd.read_csv('stylo_wiki_text.csv')
```

### LLM vs LLM vs wiki

```python
gpt3_stylos['model']= 'gpt3'
gpt4_stylos['model']= 'gpt4'
llama2_stylos['model']= 'llama2'
llama3_stylos['model']= 'llama3'
orca_stylos['model']= 'orca'
falcon_stylos['model']= 'falcon'
```

```python
gpt3_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
gpt4_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
llama2_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
llama3_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
orca_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
falcon_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
```

```python
wiki_stylos['model']= 'wiki'
wiki_text_stylos['model']= 'wikishort'

wiki_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
wiki_text_stylos.drop(['text','Unnamed: 0'], axis=1, inplace=True)
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

import random
import statistics as s

styles1 = [wiki_stylos, wiki_text_stylos, gpt3_stylos, gpt4_stylos, llama2_stylos, llama3_stylos, orca_stylos, falcon_stylos]
styles2 = [wiki_stylos, wiki_text_stylos, gpt3_stylos, gpt4_stylos, llama2_stylos, llama3_stylos, orca_stylos, falcon_stylos]

states = random.sample(range(0, 500), 10)

i = 0
mean_accs = []
for model1 in styles1:
    i = i + 1
    for model2 in styles2:
        print(i)
        i = i + 1
        if model1.equals(model2):            
            i = i + 1
            mean_accs.append(1.0)
            continue        
        accs = []
        for state in states:
            #print(state)
            stylo = pd.concat([model1, model2])
            stylo.dropna(inplace=True)
                
            X = stylo[['L_ADJ_COMPARATIVE', 'L_FUNC_T', 'FOS_FRONTING', 'L_TYPE_TOKEN_RATIO_LEMMAS']].to_numpy()
            X = np.nan_to_num(X)
        
            y = stylo['model'].to_numpy()

            clf = tree.DecisionTreeClassifier()
        
            X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                                test_size=0.33, 
                                                                random_state=state)
            clf.fit(X_train, y_train)
    
            predictions = clf.predict(X_test)
            acc = accuracy_score(predictions, y_test)
            accs.append(acc)
        mean_accs.append(s.mean(accs))
```
